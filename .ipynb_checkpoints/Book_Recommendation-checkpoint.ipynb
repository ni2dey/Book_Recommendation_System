#popularity based recommendation system

```python
books_rating=books.merge(ratings,on='ISBN')
books_rating
```

```python
num_rating=books_rating.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating.rename(columns={'Book-Rating':'No_of_ratings'},inplace=True)
num_rating
```

```python
avg_rating = books_rating.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating': 'Average_ratings'}, inplace=True)
avg_rating
```

```python
popular_rating=avg_rating.merge(num_rating,on='Book-Title')
popular_rating
```

```python
popular_rating=popular_rating[popular_rating['No_of_ratings']>250]
popular_rating
```

```python
popular_df=popular_rating.sort_values('Average_ratings',ascending=False).head(50)
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Average_ratings','No_of_ratings','Publisher','Image-URL-M']]
popular_df
```

```python
import pickle
pickle.dump(popular_df,open('popular.pickle','wb'))
```

##Colaborative Filter Based Recommendation System

```python
user_num_rating=books_rating.groupby('User-ID')['Book-Rating'].count().reset_index()
user_num_rating.rename(columns={'Book-Rating':'No_of_rating'},inplace=True)
user_num_rating
```

```python
a=user_num_rating[user_num_rating['No_of_rating']>200]
a
```

```python
user=np.array(a['User-ID'])
```

```python
filtered_rating=books_rating[books_rating['User-ID'].isin(user)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title')['Book-Rating'].count().reset_index()
filtered_book=np.array(y[y['Book-Rating']>=50]['Book-Title'])
len(filtered_book)
```

```python
filtered_df=filtered_rating[filtered_rating['Book-Title'].isin(filtered_book)]
filtered_df.drop_duplicates()
```

```python
df=filtered_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
df
```

```python
df=df.fillna(0)
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(df)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
def recommend(book):
    index=np.where(df.index==book)[0][0]
    book_list=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in book_list:
        recommend=df.index[i[0]]
        print(recommend)
```

```python
recommend("Zoya")
```
